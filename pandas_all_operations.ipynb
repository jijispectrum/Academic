```python
#functions
# importing pandas module
import pandas as pd 

# Define a dictionary containing employee data 
data3 = {'Name':['Jai', 'Anuj', 'Jai', 'Princi', 
				'Gaurav', 'Anuj', 'Princi', 'Abhi'], 
		'Age':[27, 24, 22, 32, 
			33, 36, 27, 32], 
		'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj',
				'Jaunpur', 'Kanpur', 'Allahabad', 'Aligarh'], 
		'Qualification':['Msc', 'MA', 'MCA', 'Phd',
						'B.Tech', 'B.com', 'Msc', 'MA']} 
	

# Convert the dictionary into DataFrame 
df7 = pd.DataFrame(data3)

print(df7) 
```

```python
# using keys for grouping
# data

df7.groupby('Name').groups
```

```python
# iterating an element
# of group containing 
# multiple keys

grp = df7.groupby(['Name', 'Qualification'])
for name, group in grp:
	print(name)
	print(group)
	print()
```

```python
# selecting a single group

grp = df7.groupby('Name')
grp.get_group('Jai')
```

```python
# selecting object grouped
# on multiple columns

grp = df7.groupby(['Name', 'Qualification'])
grp.get_group(('Jai', 'Msc'))
```

```python
# performing aggregation using
# aggregate method
import numpy as np
grp1 = df7.groupby('Name')

grp1.aggregate(np.sum)

```

```python
# performing aggregation on
# group containing multiple
# keys
grp1 = df7.groupby(['Name', 'Qualification'])

grp1.aggregate(np.sum)
```

```python
# applying a function by passing
# a list of functions

grp = df7.groupby('Name')

grp['Age'].agg([np.sum, np.mean, np.std])
```

```python
#Grouping rows in pandas
# importing Pandas 
import pandas as pd 

# example dataframe 
example = {'Team':['Arsenal', 'Manchester United', 'Arsenal', 
				'Arsenal', 'Chelsea', 'Manchester United', 
				'Manchester United', 'Chelsea', 'Chelsea', 'Chelsea'], 
					
		'Player':['Ozil', 'Pogba', 'Lucas', 'Aubameyang', 
					'Hazard', 'Mata', 'Lukaku', 'Morata', 
										'Giroud', 'Kante'], 
										
		'Goals':[5, 3, 6, 4, 9, 2, 0, 5, 2, 3] } 

df8 = pd.DataFrame(example) 

print(df8) 
```

```python
total_goals = df8['Goals'].groupby(df8['Team']) 

# printing the means value 
print(total_goals.mean())	 
```

```python
#combining multiple rows
# importing pandas as pd 
import pandas as pd 

# Creating a dictionary 
d = {'id':['1', '2', '3'], 
	'Column 1.1':[14, 15, 16], 
	'Column 1.2':[10, 10, 10], 
	'Column 1.3':[1, 4, 5], 
	'Column 2.1':[1, 2, 3], 
	'Column 2.2':[10, 10, 10], } 

# Converting dictionary into a data-frame 
df = pd.DataFrame(d) 
print(df) 
```

```python
# Creating the groupby dictionary 
groupby_dict = {'Column 1.1':'Column 1', 
				'Column 1.2':'Column 1', 
				'Column 1.3':'Column 1', 
				'Column 2.1':'Column 2', 
				'Column 2.2':'Column 2' } 

# Set the index of df as Column 'id' 
df = df.set_index('id') 

# Groupby the groupby_dict created above 
df = df.groupby(groupby_dict, axis = 1).min() 
print(df) 
```

```python
# importing pandas module
import pandas as pd 
 
# Define a dictionary containing employee data 
data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']} 
   
# Define a dictionary containing employee data 
data2 = {'Name':['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'], 
        'Age':[17, 14, 12, 52], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
 
# Convert the dictionary into DataFrame  
df = pd.DataFrame(data1,index=[0, 1, 2, 3])
 
# Convert the dictionary into DataFrame  
df1 = pd.DataFrame(data2, index=[4, 5, 6, 7])
 
print(df, "\n\n", df1) 
```

```python
# using ignore_index
res = pd.concat([df, df1], ignore_index=True)
 
res
```
